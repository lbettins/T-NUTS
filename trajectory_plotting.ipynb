```python
# Import packages
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get colors from coolwarm colormap
colors = plt.get_cmap('cividis', len(phi))
```

```python
# Set filename
filename = "trajectories_14.txt"
filename = "/Users/lancebettinson/scratch/nuts_out/trajectories_17.txt"
# Open file and populate data structure
phi = []
E = []
grad = []
with open(filename,'r') as f:
    line = f.readline()
    line = f.readline()
    while line:
        split = line.split('\t')
        phi_ = split[0].strip('\\[').strip('\\]').split(' ')
        while '' in phi_:
            phi_.remove('')
        phi.append([float(p) for p in phi_])
        #phi.append(np.array(split[0]))
        E.append(float(split[1]))
        line = f.readline()
    f.close()
```

```python
# Convert to degrees
phi = np.array(phi)
phi = phi*180/np.pi
```

```python
# Change matplotlib backend
%matplotlib notebook
# Create figure and add axes
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
```
